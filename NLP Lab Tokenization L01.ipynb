```python
import nltk
text = "NLTK is a great NLP toolkit. It makes processing text easy! Natural language processing primarily deals with the processing of human languages using ml and dl"
sentences = nltk.word_tokenize(text)
print(sentences)
```

```python
text = "Hello World!"
words = nltk.word_tokenize(text)
print("Characters in the text:")
for word in words:
    for char in word:
        print(char)
```

```python
tokenizer = nltk.WordPunctTokenizer()
tokens=tokenizer.tokenize("Don't split contractions. E-mail: hello@example.com!")
print(tokens)
```

```python
tokenizer = nltk.TreebankWordTokenizer();
tokens = tokenizer.tokenize("Have a look at NLTK's tokenizer's.")
print(tokens)
```

```python
tokenizer = nltk.RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize("Custom rule.keep only words & numbers,drop punctuation!")
print(tokens)
```
